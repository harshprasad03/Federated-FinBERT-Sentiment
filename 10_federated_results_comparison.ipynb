```python
import pandas as pd

fedavg_df = pd.read_csv("results/fedavg/fedavg_10_rounds.csv")
fedprox_df = pd.read_csv("results/fedprox/fedprox_mu_0.05.csv")
adaptive_df = pd.read_csv("results/adaptive/adaptive_fedavg_10_rounds.csv")

fedavg_df.head(), fedprox_df.head(), adaptive_df.head()
```

```python
#Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.plot(fedavg_df["Round"], fedavg_df["Avg_F1"], marker="o", label="FedAvg (10)")
plt.plot(fedprox_df["Round"], fedprox_df["Avg_F1"], marker="s", label="FedProx (=0.05)")
plt.plot(adaptive_df["Round"], adaptive_df["Avg_F1"], marker="^", label="Adaptive FedAvg")
plt.xlabel("Federated Round")
plt.ylabel("Average F1-score")
plt.title("Global Model Convergence Across Algorithms")
plt.grid(True)
plt.legend()
plt.show()
```

```python
summary = pd.DataFrame({
    "Method" : [
        "FedAvg (10 rounds)",
        "FedProx (=0.05)",
        "Adaptive FedAvg"
    ],
    "Final Avg F1" : [
        fedavg_df["Avg_F1"].iloc[-1],
        fedprox_df["Avg_F1"].iloc[-1],
        adaptive_df["Avg_F1"].iloc[-1]
    ]
})

print(summary)
```
